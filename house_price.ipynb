```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
house_data = {
    "Square_Feet": [1200, 1500, 1800, 2100, 2400, 2700, 3000, 3300, 3600, 3900,
                    1400, 1700, 2000, 2300, 2600, 2900, 3200, 3500, 3800, 4100],
    "House_Price": [245, 310, 355, 425, 480, 520, 595, 650, 715, 780,
                    290, 340, 410, 465, 510, 580, 630, 700, 760, 820]
}
df = pd.DataFrame(house_data)
df
```

```python
X = df[['Square_Feet']] #Independent Variable
y = df['House_Price']#dependent Variable
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y, test_size=0.3,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
m = LinearRegression()
m.fit(X_train,y_train)
```

```python
y_predt = m.predict(X_test)
y_predt
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_predt)
mse
```

```python
y_reg = (m.coef_)*X + m.intercept_
```

```python
plt.scatter(X,y,color = '#C70039',marker='o')
plt.plot(X, y_reg, color='k', linewidth=1, label="Predicted Prices")
plt.xlabel('Square ft.')
plt.ylabel('House Price')
plt.title('Linear Regression')
plt.legend()
plt.grid()
```
